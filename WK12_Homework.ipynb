Models made from both unsupervised and supervised learning should be thoroughly trained on existing data and tested on existing data that is previously unseen by the model to ensure that the results are accurate in both precision and recall and that the results of the model are actionable for the business. ### Pre-processing #2, reduce number of categories

```python
#categorize job titles based on related roles to reduce overall number of categories

lending_club.loc[lending_club['emp_title'].str.contains
             ('director', na=False, case=False, regex=True), 'emp_title'] = 'director'
lending_club.loc[lending_club['emp_title'].str.contains
             ('manager|operation|supervisor|senior|sr|management|foreman|sergeant|sheriff|captain|mgr|lead', na=False, case=False, regex=True), 'emp_title'] = 'supervisor'
lending_club.loc[lending_club['emp_title'].str.contains
             ('worker|builder|forklift|stock|tech|repair|mechanic|tester|attendant|asmbler|operator|install|inspector|toolmaker|maintenance', na=False, case=False, regex=True), 'emp_title'] = 'technician'
lending_club.loc[lending_club['emp_title'].str.contains
             ('consult|consultant|coordinator', na=False, case=False, regex=True), 'emp_title'] = 'consultant/coordinator'
lending_club.loc[lending_club['emp_title'].str.contains
             ('engineer', na=False, case=False, regex=True), 'emp_title'] = 'engineer'
lending_club.loc[lending_club['emp_title'].str.contains
             ('executive|chief|president|vp', na=False, case=False, regex=True), 'emp_title'] = 'executive'
lending_club.loc[lending_club['emp_title'].str.contains
             ('driver', na=False, case=False, regex=True), 'emp_title'] = 'driver'
lending_club.loc[lending_club['emp_title'].str.contains
             ('teach|education|principal', na=False, case=False, regex=True), 'emp_title'] = 'teacher'
lending_club.loc[lending_club['emp_title'].str.contains
             ('service|officer|merch|trainer|recruiter|representative|rep|cook|sales|customer|bartender', na=False, case=False, regex=True), 'emp_title'] = 'service representative'
lending_club.loc[lending_club['emp_title'].str.contains
             ('postdoc| post doctoral', na=False, case=False, regex=True), 'emp_title'] = 'postdoc'
lending_club.loc[lending_club['emp_title'].str.contains
             ('doctor|practitioner', na=False, case=False, regex=True), 'emp_title'] = 'medical practitioner'
lending_club.loc[lending_club['emp_title'].str.contains
             ('nurse|phlebot|phel|therapist|medic', na=False, case=False, regex=True), 'emp_title'] = 'medical associate'
lending_club.loc[lending_club['emp_title'].str.contains
             ('secretary|admin|examiner|resources|work comp|data entry|hr|enrollment|paralegal', na=False, case=False, regex=True), 'emp_title'] = 'administrative'
lending_club.loc[lending_club['emp_title'].str.contains
             ('lawyer', na=False, case=False, regex=True), 'emp_title'] = 'lawyer'
lending_club.loc[lending_club['emp_title'].str.contains
             ('tax|accounting|billing|financial|clerk', na=False, case=False, regex=True), 'emp_title'] = 'accounting'
lending_club.loc[lending_club['emp_title'].str.contains
             ('data|analyst', na=False, case=False, regex=True), 'emp_title'] = 'technology'
lending_club.loc[lending_club['emp_title'].str.contains
             ('social|intervention|youth', na=False, case=False, regex=True), 'emp_title'] = 'social worker'


lending_club['emp_title'].value_counts()
```

### Pre-processing #3, convert dtypes to numeric

```python
#view all datatypes
lending_club.dtypes
```

```python
#select columns with non-numeric datatypes
lending_club.select_dtypes(include=['object'])
```

```python
#use pd.factorize to encode categorical variables to numerical, starting at 1

#encode emp_title column
lending_club['emp_title_encoded'] = pd.factorize(lending_club['emp_title'])[0]+1

#compare value count of original and encoded columns
lending_club[['emp_title', 'emp_title_encoded']].value_counts()
```

```python
#encode state column
lending_club['state_encoded'] = pd.factorize(lending_club['state'])[0]+1

#compare value count of original and encoded columns
lending_club[['state', 'state_encoded']].value_counts()
```

```python
#encode homeownership column
lending_club['homeownership_encoded'] = pd.factorize(lending_club['homeownership'])[0]+1

#compare value count of original and encoded columns
lending_club[['homeownership', 'homeownership_encoded']].value_counts()
```

```python
#encode verified_income column
lending_club['verified_income_encoded'] = pd.factorize(lending_club['verified_income'])[0]+1

#compare value count of original and encoded columns
lending_club[['verified_income', 'verified_income_encoded']].value_counts()
```

```python
#encode loan_purpose column
lending_club['loan_purpose_encoded'] = pd.factorize(lending_club['loan_purpose'])[0]+1

#compare value count of original and encoded columns
lending_club[['loan_purpose', 'loan_purpose_encoded']].value_counts()
```

```python
#encode application_type column
lending_club['application_type_encoded'] = pd.factorize(lending_club['application_type'])[0]+1

#compare value count of original and encoded columns
lending_club[['application_type', 'application_type_encoded']].value_counts()
```

```python
#encode grade column, sort= True to maintin relationship between grades
lending_club['grade_encoded'] = pd.factorize(lending_club['grade'], sort=True)[0]+1

#compare value count of original and encoded columns
lending_club[['grade', 'grade_encoded']].value_counts()
```

```python
#encode sub_grade column, sort= True to maintin relationship between grades
lending_club['sub_grade_encoded'] = pd.factorize(lending_club['sub_grade'], sort=True)[0]+1

#compare value count of original and encoded columns
lending_club[['sub_grade', 'sub_grade_encoded']].value_counts()
```

```python
#encode loan_status column
lending_club['loan_status_encoded'] = pd.factorize(lending_club['loan_status'])[0]+1

#compare value count of original and encoded columns
lending_club[['loan_status', 'loan_status_encoded']].value_counts()
```

```python
#encode initial_listing_status column
lending_club['initial_listing_status_encoded'] = pd.factorize(lending_club['initial_listing_status'])[0]+1

#compare value count of original and encoded columns
lending_club[['initial_listing_status', 'initial_listing_status_encoded']].value_counts()
```

```python
#encode disbursement_method column
lending_club['disbursement_method_encoded'] = pd.factorize(lending_club['disbursement_method'])[0]+1

#compare value count of original and encoded columns
lending_club[['disbursement_method', 'disbursement_method_encoded']].value_counts()
```
