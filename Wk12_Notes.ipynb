```python
jeopardy_df = pd.read_csv("../Week_05/JEOPARDY_CSV.csv")
jeopardy_df.head(2)
```

```python
#remove extra spaces
jeopardy_df.columns = jeopardy_df.columns.str.replace(" ", "")
jeopardy_df.head(2)
```

```python
jeopardy_df['AirDate'] = pd.to_datetime(jeopardy_df['AirDate'],format='%m/%d/%Y')
jeopardy_df.dtypes
```

```python
#assign is typiucally used to assign a new column
jeopardy_df = jeopardy_df.assign(Month=lambda x: x['AirDate'].dt.month)
jeopardy_df.head(2)
```

```python
#how do we do the same thing with .apply()
jeopardy_df['Year'] = jeopardy_df.apply(lambda x: x['AirDate'].year, axis=1)
jeopardy_df.head()

#differences in where to name new column, need to remove .dt, and add axxis arg.
```

#### Map and Reduce

Map allows up to apply a function over all the input values, can accept various sizes of inputs: dataframe, column, lists

Reduce performs a computation or aggregation over the entire list, only accepts list

```python
#replace filter with map
mapped_list = list(map(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))
mapped_list
```

```python
from functools import reduce
reduced_list = list(reduce(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))
reduced_list

#needs to do aggregate mathematics to work
```
